```python
%sql
create catalog if not exists lakehouse;
create schema if not exists lakehouse.01_raw;
create schema if not exists lakehouse.02_bronze;
create schema if not exists lakehouse.03_silver;
create schema if not exists lakehouse.04_gold;
create volume if not exists lakehouse.01_raw.raw;
```

```python
source_details = {
    "host" : 'https://earthquake.usgs.gov',
    "port" : 443,
    "base_path" : '/earthquakes/feed/v1.0/',
    "bearer_token" : 'na'
}
```

```python
 # You can go to catalog and create a connection through UI
 # here we are creating the connection programmatically

 from databricks.sdk import WorkspaceClient
 from databricks.sdk.service import catalog

 w = WorkspaceClient()
 w.connections.create(
     name="earthquake_api_http_connection",
     connection_type=catalog.ConnectionType.HTTP,
     options={
         "host": source_details["host"],
         "port": source_details["port"],
         "base_path": source_details["base_path"],
         "bearer_token": source_details.get("bearer_token")
             }
 )
```
